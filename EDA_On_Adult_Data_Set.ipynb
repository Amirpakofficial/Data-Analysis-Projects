```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
data = pd.read_csv(url, names=column_names, na_values=' ?', skipinitialspace=True)
```

Explore individual variables using histograms and bar charts


```python
plt.figure(figsize=(12, 6))
sns.countplot(x='education', data=data, hue='income')
plt.xticks(rotation=45)
plt.show()
```

Bivariate Analysis using a heatmap to visualize correlations between numerical features


```python
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

Multivariate Analysis using pair plots to visualize relationships between multiple variables


```python
sns.pairplot(data, hue='income', diag_kind='kde')
plt.show()
```

Creating a new feature 'age_group' based on age ranges


```python
bins = [17, 25, 35, 45, 60, 100]
labels = ['17-25', '26-35', '36-45', '46-60', '60+']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)
```

Detect outliers using box plots for numerical features


```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='education_num', y='age', data=data, hue='income')
plt.title('Education Num vs Age grouped by Income')
plt.show()
```

Data transformation using log transformation on 'capital_gain' and 'capital_loss'

```python
data['log_capital_gain'] = data['capital_gain'].apply(lambda x: 0 if x == 0 else np.log(x))
data['log_capital_loss'] = data['capital_loss'].apply(lambda x: 0 if x == 0 else np.log(x))
```

Summary statistics and insights


```python


income_counts = data['income'].value_counts()
age_group_counts = data['age_group'].value_counts()

print("Income counts:")
print(income_counts)
print("\nAge Group counts:")
print(age_group_counts)
```

Visualizing the distribution of income across age groups

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', data=data, hue='income', order=labels)
plt.title('Income Distribution across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```
